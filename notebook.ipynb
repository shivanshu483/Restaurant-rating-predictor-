## Filtering the data frame:

```python
df = df[df['Rating text']!= 'Not rated']
```

```python
plt.figure(figsize=(6, 4))  
df['Rating text'].value_counts().plot(kind = 'pie',startangle=90,wedgeprops={'edgecolor': 'black'})
plt.title("Average Rarings")
plt.ylabel('')
plt.show()
```

```python
x = df[['Average Cost for two', 'Has Table booking','Has Online delivery','Price range']]
```

```python
# Above 2 below 2.5 Poor
# Above 2.5 below 3.5 Avarege 
# Above 3.5 below 4.0 Good
# Above 4 below 4.5 Very Good
# Above 4.5 Excellent
```

```python
y = df[['Aggregate rating']]
```

```python
x
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
x['Has Table booking'] = lb.fit_transform(x['Has Table booking'])                                                     
```

```python
lb.classes_
```


Has Table booking : 1 is yes , 0 is no

```python
x['Has Online delivery'] = lb.fit_transform(x['Has Online delivery'])
```

```python
lb.classes_
```

Has Online delivery : 1 is yes , 0 is no

## Sceling

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
```

```python
x = sd.fit_transform(x)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def modelresult(prediction):
    print("Mean absoult error on model is {}".format(mean_absolute_error(y_test,prediction)))
    print("Root mean squred error on model is {}".format(np.sqrt(mean_squared_error(y_test,prediction))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.predict(x_test)
```

```python
predictionsformlr = lr.predict(x_test)
```

```python
modelresult(predictionsformlr)
```

```python
predictionsformlr
```

## Support vector regression 

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV    
```

```python
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV

# Define the SVR model
svrmodel = SVR()

# Set up the parameter grid
param_gridsvr = {
    "C": [0.1, 0.5],
    "kernel": ["linear", "poly"],
    "degree": [2]
}

# Set up and run the grid search
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
gridsvr.fit(x_train, y_train)
```

```python
predsgridsvr = gridsvr.predict(x_test)
```

```python
modelresult(predsgridsvr)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid  = {
    "max_depth" : [10,30],
    "min_samples_leaf" : [1,2],
    "min_samples_split" : [2,5]
}
```

```python
treemodel = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator = treemodel,param_grid=param_grid)
```

```python
grid_tree.fit(x_train,y_train)
```

## Prediction

```python
treepredction = grid_tree.predict(x_test)
```

```python
modelresult(treepredction)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfrmodel = RandomForestRegressor()
```

```python
param_gridrfr = {"max_depth": [5,15],"n_estimators" : [2,5,10]}
```

```python
gridrfr = GridSearchCV(rfrmodel,param_gridrfr)
```

```python
gridrfr.fit(x_train,y_train)
```

```python
randomforestpredictions = gridrfr.predict(x_test)
```

```python
modelresult(randomforestpredictions)
```

```python
gridrfr.best_params_
```

```python
grid_tree.best_params_
```

```python
from sklearn.neighbors import KNeighborsRegressor

knn = KNeighborsRegressor()
knn_param_grid = {"n_neighbors": [9,11,14,15,16]}
knn_grid_search = GridSearchCV(knn, knn_param_grid)
```
